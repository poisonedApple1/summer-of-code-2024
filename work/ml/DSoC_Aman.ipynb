```python
import pandas as pd
```

```python
df= pd.read_csv("PS_20174392719_1491204439457_log.csv")
```

```python
# Get one hot encoding of columns B
one_hot = pd.get_dummies(df['type'])
# Drop column B as it is now encoded
df = df.drop('type',axis = 1)
# Join the encoded df
df = df.join(one_hot)
df
```

```python
df=df.drop('nameOrig',axis=1)
df
```

```python
df['dest_type']=df['nameDest'].str[0]
```

```python
df
```

```python
df=df.drop('nameDest', axis=1)
```

```python
# Get one hot encoding of columns B
one_hot = pd.get_dummies(df['dest_type'])
# Drop column B as it is now encoded
df = df.drop('dest_type',axis = 1)
# Join the encoded df
df = df.join(one_hot)
df
```

```python
df=df.drop('M',axis=1)
```

```python
df
```

```python
def norm_column(df,col):
    df_col=df[col]
    min_val=df_col.min()
    max_val=df_col.max()
    df[col]=(df_col-min_val)/(max_val-min_val)
```

```python
norm_cols ={
    "step",
    "amount",
    "oldbalanceOrg",
    "newbalanceOrig",
    "oldbalanceDest",
    "newbalanceDest"
}
for col in norm_cols:
  norm_column(df,col)
```

under sampling to manage class imbalance b/w fraud and not fraud

```python
df_isFraud=df[df["isFraud"]==1]
```

```python
df_isNotFraud=df[df["isFraud"]==0]
```

```python
df_isFraud
```

```python
df_isNotFraud=df_isNotFraud.sample(n=8213)
```

```python
df_isNotFraud
```

```python
df=pd.concat([df_isFraud,df_isNotFraud])
```

```python
df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
```

```python
from sklearn.tree import export_graphviz
from IPython.display import Image
import graphviz
```

```python
X = df.drop('isFraud', axis=1)
y = df['isFraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
